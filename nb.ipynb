```python
import random

LETTER_POOL = {
    'A': 9, 
    'B': 2, 
    'C': 2, 
    'D': 4, 
    'E': 12, 
    'F': 2, 
    'G': 3, 
    'H': 2, 
    'I': 9, 
    'J': 1, 
    'K': 1, 
    'L': 4, 
    'M': 2, 
    'N': 6, 
    'O': 8, 
    'P': 2, 
    'Q': 1, 
    'R': 6, 
    'S': 4, 
    'T': 6, 
    'U': 4, 
    'V': 2, 
    'W': 2, 
    'X': 1, 
    'Y': 2, 
    'Z': 1
}


SCORE_CHART = {
    'A': 1, 
    'B': 3, 
    'C': 3, 
    'D': 2, 
    'E': 1, 
    'F': 4, 
    'G': 2, 
    'H': 4, 
    'I': 1, 
    'J': 8, 
    'K': 5, 
    'L': 1, 
    'M': 3, 
    'N': 1, 
    'O': 1, 
    'P': 3, 
    'Q': 10, 
    'R': 1, 
    'S': 1, 
    'T': 1, 
    'U': 1, 
    'V': 4, 
    'W': 4, 
    'X': 8, 
    'Y': 4, 
    'Z': 10
}

def draw_letters():
    # create letters list
    letters = []
    # create pool dictionary by copying LETTER_POOL dicitonary
    pool = LETTER_POOL.copy()
    # set loop to iterate while the number of letters in letters list is less than 10
    while len(letters) < 10:
        # create hand list  
        # turn pool dictionary into a list
        # get random character string from pool list
        hand = random.choice(list(pool))
        # reference pool dictionary value at a random index in the hand list
        # iterate through the rest of the loop if the pool value is <= 0
        if pool[hand] <= 0:
            # exclude pool values <= 0 from being appended to letters list
            continue
        # subtract 1 from the key value in pool dictionary at the random index
        pool[hand] -= 1
        # create string array by appending the letters list to the hand list
        letters.append(hand)
    # return the letters list which is now a string array
    return letters

def uses_available_letters(word, letter_bank):
    # make bank dictionary by copying letter_bank dictionary
    bank = letter_bank.copy()
    # declare character variable in for loop to represent a letter
    for character in word:
        # make character uppercase
        character = character.upper()
        if character in bank:
            # when there is a letter in the bank dicitonary remove that character from the bank
            bank.remove(character)
        else:
            # return false if the character is not in bank
            return False
    # return true if the the word meets the function condition (is in bank)
    return True          


def score_word(word):
    # set score to counter to 0
    score = 0
    for character in word:
        # make the letter uppercase if it's in word
        character = character.upper() 
        # add the value of the letter by referencing the SCORE_CHART dictionary at that letter
        score += SCORE_CHART[character] 
    # check if the length of word is >= 7
    if len(word) >= 7:
        # add an extra 8 to the score if length of word >= 7
        score += 8
    # return the score
    return score


def get_highest_word_score(word_list):
    # candidate = (None, 0)
    # for word in word_list:
    #     score_word(word)

    candidates = [(w, score_word(w)) for w in word_list]

    scores = [score_word(w) for w in word_list]







# def get_highest_word_score(word_list):
#     # create word list
#     word_scores = []
#     # check word in word_list
#     for word in word_list:
#         # add the word and the score of the word to the empty word_scores list
#         # access the score of the word by calling score word function at word  
#         word_scores.append((word, score_word(word)))
#     # sort tuple integer values from highest to lowest     
#     word_scores.sort(key=lambda x: x[1], reverse = True)

#     # create a variable for tied highest score
#     # find tied highest score by is found and saved to
#         # tied highest score variable
#         # by indexing the word scores list strings and integers
#     tied_highest_score = word_scores[0][1]
#     # create empty list for tied highest scores
#     tied_highest_scores = []
#     # check if the word score is a tied highest score
#     for word_score in tied_highest_score:
#         # if tied highest score is the same as the word score (the value, indicated by indexing the interger value of the string at [1])
#         if tied_highest_score == word_score[1]:
#             # add that word score to the tied_highest_scores list
#             tied_highest_scores.append(word_score)
#     # if the there is just one highest score in the tied_highest_scores list
#     if len(tied_highest_scores) == 1:
#         # return tuple string and tuple value
#         return word_scores[0]

#     # tied_highest_scores.sort(key=lambda x: len(x[0]), reverse = True)
#     # if len(tied_highest_scores[0][0]) == 1:

```
